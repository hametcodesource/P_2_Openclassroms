## Imputation de notre table à l'aide des methodes Backwardfill et ForwardFill

Le valeurs maquante sur la table correspondent aux anneés ou une indicateur n'a pas été etudié. de ce fait 
ces methodes permettent de changer les valeures maquantes par celles qui sont plus recentes.
     
      ForwardFill: Remplissage vers l'avant des valeurs manquantes imputées à l'aide de la dernière valeur
                   observée.
                   Exemple: [90.50,Nan,20.9]=====>[90.50,20.9,20.9]
      Backwardfill: Le remplissage en arrière signifie remplir les valeurs manquantes avec le point de 
                    données suivant.
                   Exemple: [Nan,Nan,20]=====> [20,20,20]


```python
dataset=EdStatsData.loc[EdStatsData['Indicator Name'].isin(List_indicator),:]

dataset=dataset.T

dataset.iloc[4:,:]=dataset.iloc[4:,:].backfill(axis="rows")
dataset.iloc[4:,:]=dataset.iloc[4:,:].ffill(axis="rows")

```

On voit que la valeur des indicateurs ne sont pas toute au meme échelle, ainsi pour remédier a se probleme on  va normaliser le jeu de donnée grace à la fonction suivante.

```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result
```

```python
dataset_pays=normalize(dataset_pays)
dataset_pays
```

## Etudions de correlation des indicateurs par la methode de pearson

Le coefficient de Pearson est un indice reflétant une relation linéaire entre deux variables continues. Le coefficient de corrélation varie entre -1 et +1, 0 reflétant une relation nulle entre les deux variables, une valeur négative (corrélation négative) signifiant que lorsqu'une des variable augmente, l'autre diminue ; tandis qu'une valeur positive (corrélation positive) indique que les deux variables varient ensemble dans le même sens

```python
corr_indic=dataset_pays.corr()
corr_indic
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
mask=np.triu(np.ones_like(corr_indic,dtype=bool))
sns.heatmap(corr_indic,mask=mask,center=0,cmap='RdBu',linewidths=1,annot=True,fmt=".2f",vmin=-1,vmax=1,ax=ax)
plt.title("carte de corrélation",fontsize=15,fontweight='bold')
plt.show()
```

Le graphe nous montre une correlation positive ou negative entre les indicateurs, d'une part  plus la coleur passe vers le bleu plus la corréllation entre les indicateurs est positives, d'autre part plus la couleur tend vers la coleur rouge plus la correllation est negatives.

```python
corr_matrix = corr_indic.abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
to_drop_corr = [column for column in upper.columns if any(upper[column] > 0.50)] 
```

```python
dataset_pays.drop(to_drop_corr, axis=1, inplace=True)
```

Dans cette étude on a retenue les indiacteurs dont les coefficient de correllation sont inferieur a 0.50. ainsi grace à la methode de pearson on obtient une table reduit a 6 indiacteurs.

## Pour aller plus loin etudions l'independance des indicateurs par la methode des p_values

Pour déterminer si les variables sont indépendantes, comparez la valeur de p au seuil de signification. En général, un seuil de signification (noté alpha ou α) de 0,05 fonctionne bien. Un seuil de signification de 0,05 indique 5 % de risque de conclure à tort qu'il existe une association.


Valeur de p ≤ α : les variables présentent une association statistiquement significative (rejeter H0)
Si la valeur de p est inférieure ou égale au seuil de signification, vous pouvez rejeter l'hypothèse nulle et conclure qu'il existe une association statistiquement significative entre les variables.


Valeur de p > α : impossible de conclure que les variables sont associées (ne pas rejeter H0)
Si la valeur de p est supérieure au seuil de signification, vous ne pouvez pas rejeter l'hypothèse nulle, car vous ne possédez pas suffisamment de preuves pour conclure que les variables sont associées.

```python
from scipy.stats.stats import pearsonr
a = np.empty((len(dataset_pays.columns),len(dataset_pays.columns),))
a[:] = np.nan
for i in range(0,len(dataset_pays.columns)):
    for j in range(0,len(dataset_pays.columns)):
        a[i,j] = pearsonr(dataset_pays.iloc[:,i], dataset_pays.iloc[:,j])[1]

df_pvalue = round(pd.DataFrame(a, columns=dataset_pays.columns, index = dataset_pays.columns),5)
```

```python
cm = sns.light_palette("green", as_cmap=True) 

df_pvalue.\
style.background_gradient(cmap=cm).set_precision(2)
```

Le tableau des P_valeus nous prouve que pour un seuil de 5%:


      le  taux d'accée a internet et le taux net d'etudiant en mobilité internationale
       sont non independant,


      le taux d'accée a internet  et le taux de non escolarisation sont  non independant,


      le taux d'accée a internet et le taux d'abandon cululé   sont non independant

```python
fig, ax = plt.subplots(figsize=(10, 5))
mask=np.triu(np.ones_like(df_pvalue,dtype=bool))
sns.heatmap(df_pvalue,mask=mask,center=0,cmap='RdBu',linewidths=1,annot=True,fmt=".2f",vmin=-1,vmax=1,ax=ax)
plt.title("P_value",fontsize=15,fontweight='bold')
plt.show()
```

```python
p_value_matrix = df_pvalue.abs()

upper = p_value_matrix.where(np.triu(np.ones(p_value_matrix.shape), k=1).astype(np.bool))
to_drop_p_values = [column for column in upper.columns if any(upper[column]<= 0.05)]
to_drop_p_values
```

Grace à cette methode des p_values on a en trois indicateurs qui son non independant pour un seuil de 5%.ainsi  notre table de 6 indicateurs est reduit a 3 indcateurs ce qui nous facilité l'analyse de ces dernierres.

```python
dataset_pays.drop(to_drop_p_values, axis=1, inplace=True)
```

## Evolution des indicateurs au cour du temps pour les pays avec un fort potentiel


La fonction plot_indic  permet de tracer l'evoultion des indicateurs dans le temps pour un pays donnée.

```python
def plot_indic(pays,indicateur,dataset,ax=None):
    dataset=dataset.loc[(dataset['Country Name']==pays)& (dataset['Indicator Name']==indicateur),:]
    if dataset.shape[0]==0:
        print(f'Error!!: indicateur non etudié dans ce pays: {pays}')
        print(dataset)
    else:
        
        fig, ax = plt.subplots(figsize=(40, 10))
        for i in range(dataset.shape[0]):
            plt.plot(dataset.iloc[i,2:],label=dataset.iloc[i,1])
        plt.title(f'Evelution des indicateur au {dataset.iloc[0,0]}')
        plt.legend()
        plt.show()

        
#plot_indic('Italy',"Internet users (per 100 people)",Project_data)
```
